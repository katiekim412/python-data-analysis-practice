```python
def stock_price(ticker):
    url = f'https://finance.naver.com/item/main.naver?code={ticker}'
    res = requests.get(url = url)
    soup = BTS(markup = res.text, features = 'html.parser')
    items = soup.select('#rate_info_krx > div > p.no_today > em > span')
    return items[0].text
```

```python
stock_price(ticker = '005930')
```

```python
class MyBTS(BTS):
    def html_text(self, css):
        items = self.select(css)
        return [item.text for item in items]
    def html_attr(self, css, key = 'href'):
        items = self.select(css)
        return [item[key] for item in items]
```

```python
soup2 = MyBTS(markup = res.text, features = 'html.parser')
```

```python
soup2.html_text(css = '.articleSubject > a')
```

```python
soup2.html_attr(css = '.articleSubject > a')
```

```python
soup2.html_text(css = 'span.press')
```

```python
soup2.html_text(css = 'span.wdate')
```

```python
def apt_price(areaCd, ymonth, rows = 9999, page = 1):
    url = 'http://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade'
    api_key = ''
    query = {
        'LAWD_CD': areaCd, 
        'DEAL_YMD': ymonth, 
        'serviceKey': api_key, 
        'numOfRows': rows, 
        'pageNo': page
    }
    res = requests.get(url = url, params = query)
    dat = xmltodict.parse(res.text)
    df = pd.DataFrame(data = dat['response']['body']['items']['item'])
    return df
```

```python
df = apt_price(areaCd = '41630', ymonth = '202504')
df.shape
```

```python
ymonths = list(range(202401, 202413))
print(ymonths)
```

```python
from tqdm.notebook import tqdm
```

```python
dfs = pd.DataFrame()
for ymonth in tqdm(ymonths):
    df = apt_price(areaCd = '41630', ymonth = ymonth)
    dfs = pd.concat(objs = [dfs, df])
```
